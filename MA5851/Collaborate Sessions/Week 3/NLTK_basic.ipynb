```python
#N-gram

from nltk import FreqDist
from nltk.util import ngrams    
def compute_freq(input_text,n):
    bigramfdist = FreqDist()
    bigrams = ngrams(input_text, n)
    bigramfdist.update(bigrams)
    return bigramfdist
output = compute_freq(text1, 4)
output.most_common(10)

```

```python
from nltk.corpus import stopwords

stop_words = stopwords.words('english')
#set(stop_words)

word_tokens = text1.tokens

strip_punc = ' '.join((x for x in word_tokens if x not in [",",".","-","?",":",".\"","?\"",";","'",",\"",'"',"!","!\""])).lower().split()
#print(strip_punc)

filtered_sentence = [w for w in strip_punc if not w in stop_words]

#print(filtered_sentence)
        
output = compute_freq(filtered_sentence, 3)
output.most_common(10)
```
