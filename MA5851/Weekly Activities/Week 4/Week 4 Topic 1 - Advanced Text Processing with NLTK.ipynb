##### Chunking: sentence segmentation - Exploring text corpora

```python
import nltk

cp = nltk.RegexpParser('NOUNS: {<N.*>{4,}}')
brown = nltk.corpus.brown
for sent in brown.tagged_sents():
    tree = cp.parse(sent)
    for subtree in tree.subtrees():
        if subtree.label() == 'NOUNS':
            print(subtree)
```
